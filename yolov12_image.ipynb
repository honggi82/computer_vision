```python
from ultralytics import YOLO
```

```python
# 1. yolov12 모델 로드
model = YOLO('yolo12n.pt')

# 2. 이미지 추론 (Inference)
image = '/content/gdrive/MyDrive/Computer_Vision/bus.jpg'
results = model(image)

# 3. 결과 확인 및 저장
for result in results:
    result.show()
    # 결과 이미지 저장
    result.save(filename='result_detection.jpg')
```
