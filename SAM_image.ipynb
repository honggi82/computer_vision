```python
import cv2
from google.colab.patches import cv2_imshow
from ultralytics import SAM
```

```python
from ultralytics import SAM

# 1. SAM2  
# Ultralytics sam2_b.pt (base), sam2_l.pt (large)  .
model = SAM('sam2_b.pt')

# 2.   (Inference)
image = '/content/gdrive/MyDrive/Computer_Vision/bus.jpg'
results = model(image)

# 3.    
result1 = results[0].plot(masks=False) # bounding box
result2 = results[0].plot(boxes=False, labels=False, conf=False) # segmentation
cv2_imshow(result1)
cv2_imshow(result2)

#   
cv2.imwrite('result_boxes_only.jpg', result1)
cv2.imwrite('result_seg_only.jpg', result2)
```
