```python
import cv2
from google.colab.patches import cv2_imshow
from ultralytics import YOLO
```

```python
model = YOLO("yolo11n-pose.pt")

# 2. 이미지 추론 (Inference)
image = '/content/gdrive/MyDrive/Computer_Vision/bus.jpg'
results = model(image)
result = results[0].plot(boxes=False, labels=False, conf=False)
cv2_imshow(result)


# 결과 이미지 저장
cv2.imwrite('result_pose_detection.jpg', result)
```
